```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'flower']
df = pd.read_csv('iris.data.txt', header=None, names=cols)
df
```

```python
# from sklearn.preprocessing import LabelEncoder
# LE = LabelEncoder()
# df['flower'] = LE.fit_transform(df['flower'])
```

```python
df['flower'].unique()
```

```python
mapping={"Iris-setosa":0,"Iris-virginica":1,"Iris-versicolor":2}
df['flower']=df['flower'].apply(lambda x:mapping[x])
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['flower']), df['flower'],random_state=6,test_size=0.3)
```

```python
from sklearn.svm import SVC
SVCClassifier = SVC(kernel = 'linear' , C = 15, max_iter=5)
SVCModel = SVCClassifier.fit(x_train,y_train)
```

```python
predictions=SVCModel.predict(x_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

print("Classification Report")
print(classification_report(y_test, predictions))
# print(accuracy_score(y_test, predictions))
```

```python
array=confusion_matrix(y_test, predictions)
```

```python
l=len(df['flower'].unique())
l
```

```python
TP=[]
for i in range(l):
    TP.append(array[i][i])
TP
```

```python
FN=[]
for k in range(l):
    m=0
    for i in range(l):
        if k!=i:
            m=m+array[k][i]
    FN.append(m)
```

```python
FN
```

```python
FP=[]
for k in range(l):
    m=0
    for i in range(l):
        if k!=i:
            m=m+array[i][k]
    FP.append(m)
```

```python
FP
```

```python
TN=[]
for k in range(l):
    m=0
    for i in range(l):
        for j in range(l):
            if i!=k and j!=k:
                m=m+array[i][j]
    TN.append(m)
```

```python
TN
```

```python
recall=[]

for i in range(l):
    m=0
    m=TP[i]/(TP[i]+FN[i])
    recall.append(m)
recall

for j in range(l):
    print("Recall for ",j)
    print(recall[j])
```

```python
pre=[]

for i in range(l):
    m=0
    m=TP[i]/(TP[i]+FP[i])
    pre.append(m)
pre

for j in range(l):
    print("Precision for ",j)
    print(pre[j])
```

```python
f1=[]
for i in range(l):
    m=0
    m=2*pre[i]*recall[i]/(pre[i]+recall[i])
    f1.append(m)
f1

for j in range(l):
    print("f-score for ",j)
    print(f1[j])
```

```python
m=0
for i in range(l):
    m=m+pre[i]
m1=m/l
print("Macro-avg for precision",m1)
```

```python
m=0
for i in range(l):
    m=m+recall[i]
m2=m/l
print("Macro-avg for recall",m2)
```

```python
m=0
for i in range(l):
    m=m+f1[i]
m1=m/l
print("Macro-avg for f-score",m1)
```
